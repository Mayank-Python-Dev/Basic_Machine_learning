```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression as lr
```

```python
data = pd.read_csv('Prices.csv')
data
```

```python
#https://www.geeksforgeeks.org/ml-dummy-variable-trap-in-regression-models/
#click on this link to understand what is dummy variable and dummy variable trap 
```

```python
dummies = pd.get_dummies(data.Colony)
```

```python
dummies
```

```python
# so now we have to concatenate this attributes with the original one 
# but we have to drop one column to avoid dummy variables trap
```

```python
concat = pd.concat([data,dummies],axis = 'columns')
```

```python
concat
```

```python
final_data = concat.drop(['Colony','Pratap_Nagar'],axis='columns')
```

```python
final_data
```

```python
#drop the dependent variable Price
x = final_data.drop(['Price'],axis='columns')
```

```python
y = final_data.Price
```
