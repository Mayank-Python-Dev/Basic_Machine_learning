```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('Prices.csv')
df
```

```python
le = LabelEncoder() #create class object
df.Colony = le.fit_transform(df.Colony)
df
```

```python
Ohe = OneHotEncoder() #create class object for OneHotEncoder
```

```python
X = Ohe.fit_transform(df.Colony.values.reshape(-1,1)).toarray()
X
```

```python
dfOneHot = pd.DataFrame(X, columns = ["Colony_"+str(int(i)) for i in range(df.shape[1])])
```

```python
x =pd.concat([df, dfOneHot], axis=1)
```

```python
x
```

```python
x = x.drop(['Colony'],axis = 1)
```

```python
x
```

```python
y = x.drop(['Price'],axis = 1)
```

```python
z = x.Price
```

```python
y
```

```python
z
```

```python
from sklearn import linear_model 
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(y,z)
```

```python
model.predict([[2100,0.0,0.0,1.0]])
```

```python
model.score(y,z)
```
